```python
import dl_translate as dlt

mt = dlt.TranslationModel("mbart50")
```

```python
import pandas as pd
import pickle
data = pickle.load(open(f'meld/meld_remake.pkl','rb'))
```

```python
import gc,torch
for i in range(len(data[3])):
    for j in range(len(data[3][i])):
        torch.cuda.empty_cache()
        gc.collect()
        data[3][i][j]=mt.translate(data[3][i][j],source="English",target="Russian")
```
